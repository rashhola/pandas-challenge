```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
Total_NP = purchase_data_df.groupby('SN')['SN'].nunique().sum()
Total_NP
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Num_uniques_items =purchase_data_df['Item ID'].nunique()
#Num_uniques_items

Number_of_purchases = purchase_data_df["Purchase ID"].count()

Average_Price = purchase_data_df["Price"].mean()

total_revenue = purchase_data_df["Price"].sum()

df = pd.DataFrame({"Number of Unique Items":[Num_uniques_items],
                           "Average Price":[Average_Price], 
                           "Number of Purchases": [Number_of_purchases], 
                           "Total Revenue": [total_revenue]})


df.style.format({"Average Price":"${:,.2f}",
                                  "Total Revenue":"${:,.2f}"})
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
Gender_count_df= purchase_data_df.groupby('Gender')['SN'].nunique()
#Gender_count_df
Percentage_counts = (Gender_count_df/Total_NP).mul(100).round(2).astype(str) + "%"
#Percentage_counts
Gender_demographics =pd.DataFrame({"Total Count": Gender_count_df,"Percentage of Players": Percentage_counts})
Gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
Purchase_count = purchase_data_df.groupby("Gender")["Purchase ID"].count()
Average_Price= purchase_data_df.groupby("Gender")["Price"].mean()
Total_Purchase_Value = purchase_data_df.groupby("Gender")["Price"].sum()
Avg_Total_Purchase_per_Person = Total_Purchase_Value / Gender_count_df

GenderDemographics = pd.DataFrame({"Purchase Count": Purchase_count, 
                                    "Average Purchase Price":Average_Price,
                                    "Average Purchase Value":Total_Purchase_Value,
                                    "Average Purchase Total per Person":Avg_Total_Purchase_per_Person})
#GenderDemographics
GenderDemographics.style.format({"Average Purchase Value":"${:,.2f}",
                                  "Average Purchase Price":"${:,.2f}",
                                  "Average Purchase Total per Person":"${:,.2f}"})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
purchase_data_df["Age"]
bins_to_cut = purchase_data_df["Age"]
bins_age = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
bins_to_fill = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','>40']


purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"],bins_age, labels=bins_to_fill)
#purchase_data_df["Age Group"]

AgeGroup = purchase_data_df.groupby("Age Group")

#To find total age count(unique)
Totalage_count = AgeGroup["SN"].nunique()
Totalage_count

#To find age %
Age_percentage =  (Totalage_count/Total_NP) *100
Age_percentage

#Table summary
Demographics_table = pd.DataFrame({"Total Count": Totalage_count, "Percentage of Players": Age_percentage})
#Demographics_table

#format the summary table
Demographics_table.style.format({"Percentage of Players":"{:,.2f}"})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#To get top spenders, we need to filter by the SN
TopSpenders = purchase_data_df.groupby('SN')
TopSpenders 
```

```python
#To get purchase per top spender
PurchasePer_TopSpenders = TopSpenders["Purchase ID"].count()

#To get average purchase price
AveragePurchasePrice_TopSpenders = TopSpenders["Price"].mean()

#To get total purchase top spenders
TotalPurchase_TopSpenders= TopSpenders["Price"].sum()

#DataFrame
Spenders = pd.DataFrame({"Purchase Count": PurchasePer_TopSpenders,
                             "Average Purchase Price": AveragePurchasePrice_TopSpenders,
                             "Total Purchase Value":TotalPurchase_TopSpenders})
Spenders
```

```python
#To get top spenders, the data has to be sorted
Big5 = Spenders.sort_values(["Total Purchase Value"], ascending=False).head()
Big5
```

```python
#Apply formatting to finish up the table
Big5.style.format({"Average Purchase Total":"${:,.2f}",
                                 "Average Purchase Price":"${:,.2f}", 
                                 "Total Purchase Value":"${:,.2f}"})
```

## Most Popular Items

```python
#Most Popular(To call out the columns to work with)
View = purchase_data_df[["Item ID", "Item Name", "Price"]]
View
```

```python
GroupName_Item = View.groupby(["Item ID","Item Name"])

ItemCount= GroupName_Item["Price"].count()

Total_purchasevalue = (GroupName_Item["Price"].sum())

Item_price = Total_purchasevalue/ItemCount
```

```python
Popularity_scale = pd.DataFrame({"Purchase Count": ItemCount, 
                                   "Item Price": Item_price,
                                   "Total Purchase Value":Total_purchasevalue})
#To see the dataFrame
Popularity_scale
```

```python
Popularity_scale_sorted = Popularity_scale.sort_values(["Purchase Count"], ascending=False).head()
Popularity_scale_sorted
```

```python
#Format it Popularity_scale_sorted
Popularity_scale_sorted.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"})
```
